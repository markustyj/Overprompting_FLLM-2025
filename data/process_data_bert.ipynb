```python
import json
import pandas as pd
```

```python
def load_csv(file_path):  
    data = pd.read_csv(file_path)  
    return data  
```

```python
def convert_csv_disk(df, path_out):
    # Convert the DataFrame to the desired format  
    formatted_data = df.to_dict(orient='records')  
    with open(path_out, 'w') as outfile:  
        for line in formatted_data:
            json.dump(line, outfile)  
            outfile.write('\n')
    return formatted_data[:1]
```

```python
def class_to_int(df):
    if 'class' in df.columns : 
        for index, class_name in enumerate (df["class"].unique() ):
            df.loc[df['class'] == class_name, "class"] = index
            
    if 'b_class' in df.columns : 
        for index, class_name in enumerate (df["b_class"].unique() ):
            df.loc[df['b_class'] == class_name, "b_class"] = index        

    return df
```
